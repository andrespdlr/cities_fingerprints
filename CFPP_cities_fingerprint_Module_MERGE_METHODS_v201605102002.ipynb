```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
plt.style.use('ggplot')
```

```python
phi_beta = pd.read_csv('phi_beta.csv')
```

```python
phi_beta.plot(kind='scatter', x='mean_phi', y='beta', color = "b", figsize=(18,12))
```

```python
phi_beta['mean_phi'].plot(kind = 'kde', figsize=(18,12), color = 'blue')
```

```python
phi_beta['beta'].plot(kind = 'kde', figsize=(18,12), color = 'blue')
```

```python
betas = []
plt.figure(figsize=(18,12))
plt.xlim((0, .1))
plt.ylim((0, .2))
plt.xlabel("Euclidean Distance")
plt.ylabel("Dijkstra Distance")  
for beta in phi_beta['beta']:
    betas.append(beta)
    plt.plot([0,1], 'r', linewidth = 3)
    plt.plot([0,beta], "RoyalBlue", linewidth = .2)

average_beta = sum(betas) / len(betas)
plt.plot([0,average_beta], "b-", linewidth = 8)
```
